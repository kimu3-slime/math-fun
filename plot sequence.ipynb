```julia
function Fib(n)
    a,b = 0,1
    for i in 1:n
        a,b = b, a+b
    end
    return a
end    
```

```julia
@time x=[Fib(n) for n in 0:50]
```

```julia
plot(x,st=:bar, label="a_n") 
```
