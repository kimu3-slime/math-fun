```julia
using PyPlot
```

```julia
sample =200
x = repeat(range(-5,5,length=sample)',sample)
y = repeat(range(-5,5,length=sample),1,sample)
 
a=1
b=0
c=0
d=1
u = a*x+b*y
v = c*x+d*y
speed = sqrt.(u.^2 + v.^2)

 
PyPlot.figure(figsize=(5,5))
PyPlot.streamplot(x, y, u, v, density=1.5, color=speed, linewidth=1)
```
