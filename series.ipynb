```julia
function f_1(N)
    s=0
    for n in 1:N
        s+= n^4
    end
    return s
end
```

```julia
x=[f_1(N) for N in 1:20]
```

```julia
bar(x)
```

```julia
function f_2(N)
    s=0
    for n in 1:N
        s+= 1/n^2
    end
    return s
end
```

```julia
x=[f_2(N) for N in 1:10^3]
```

```julia
plot(x)
```

```julia
function f_3(x,N)
    s=1
    for n in 1:N
        s+= 1/factorial(big(n)) * x^n
    end
    return s
end
```

```julia
plot(x->f_3(x,3))
```

```julia
p = plot( xlim=(-1,1),legend=:topleft)
plot!(p,x->exp(x) , label="e^x")
for N in 0:5
    plot!(p, x->f_3(x,N), label=string(N))
end
plot(p)
```

```julia
function f_4(x,N)
    s=0
    for n in 1:N
        if n % 4 ==1
            s+= 1/factorial(big(n)) * x^n
        elseif n % 4 ==3
            s+= -1/factorial(big(n)) * x^n
        else
        end
    end
    return s
end
```

```julia
plot(x->f_4(x,100))
```

```julia
p = plot( xlim=(-pi,pi),legend=:topleft)
plot!(p,x->sin(x) , label="sin")
for N in 1:2:7
    plot!(p, x->f_4(x,N), label=string(N))
end
plot(p)
```

```julia
function f_5(x,N) 
    s = 0 
    for n=1:N
        if n % 2 == 1
            s +=4/(n*pi) *sin(n*x)
        else
        end
    end 
    return s 
end 

```

```julia
p = plot(legend=:topleft, xlim=(-2*pi,2*pi),xticks=([-2pi,-pi, 0, pi, 2pi],["-2", "-", "0", "","2"]))
plot!(p,x->sign.(sin.(x)) , label="square")
for N in 1:2:5
    plot!(p, x->f_5(x,N), label=string(N))
end
plot!(p, x->f_5(x,1000), label=string(1000))
plot(p)
```
