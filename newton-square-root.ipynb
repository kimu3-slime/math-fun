```julia
f(x,p)=x^2 -p
```

```julia
df(x) =2x
```

```julia
p =2
```

```julia
#初期値の設定
x_1 =2.0
```

```julia
#近似列の値を保存する配列
x = [x_1]
```

```julia
#計算を停止させたときのステップ数の宣言
N =1
```

```julia
#ニュートン法による計算

for n in 1:100
    print(n,"  ",x[n],"  ",f(x[n],p),"\n")
    y = x[n] - f(x[n],p) / df(x[n])
    push!(x,y)    
    if abs(f(x[n],p)) < 1.0e-15
        N=n
        break
    end
end
```

```julia
e=[abs(x[n]-sqrt(p)) for n in 1:N]
```

```julia
plot(e,xlims=(1,N))
```

```julia
le =[log10(abs(x[n]-sqrt(2))) for n in 1:N] 
```

```julia
plot(le,xlims=(1,N))
```
