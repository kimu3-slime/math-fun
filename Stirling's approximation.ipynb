```julia
using Plots
gr() 
```

```julia
# 型を指定しないとオーバーフローする。BigFloatは任意精度,、デフォルトは 256 ビット。
x= BigFloat(1):BigFloat(1):BigFloat(50)
```

```julia
y1 = factorial.(x)
```

```julia
e = exp(1)
f2(x) = x^x * exp(-x+1)
y2 = f2.(x)
```

```julia
f3(x) = sqrt(2*pi) *x^(x+1/2) * exp(-x)
y3 = f3.(x)
```

```julia
#n!とスターリングの近似式をグラフで図示
plot(y1,yscale=:log10, linecolor=:black,label="n!")
plot!(y2,yscale=:log10, label="n^n * exp(-n+1)")
plot!(y3,yscale=:log10, label="sqrt(2*pi) * n^(n+1/2) * exp(-n)")
```

```julia
# n! との比の漸近挙動
z1 = y1./y2
z2 = y1./y3
plot(z1,label="n^n * exp(-n+1)")
plot!(z2,label="sqrt(2*pi) * n^(n+1/2) * exp(-n)")
```
